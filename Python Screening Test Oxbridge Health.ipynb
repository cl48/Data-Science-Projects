```python
def mfunc(tblA, tblB, join, tblA_id, tblB_id, val):
    
    import numpy as np
    import pandas as pd
    
    if np.dtype(tblA[tblA_id]) != np.dtype(tblB[tblB_id]):
        return "Error mismatch on data types for Table A id and Table B id"
    elif (tblA[tblA[tblA_id].isin(tblB[tblB_id])].shape[0] == 0 and tblB[tblB[tblB_id].isin(tblA[tblA_id])].shape[0] == 0):
        return "No ids match"
    else:
        mrg = tblA.merge(tblB, how = join, left_on = tblA_id, right_on = tblB_id, validate = val)
        return mrg[[c for c in mrg.columns if tblB_id not in c]]
```

```python
# Problem 2

import pandas as pd
import numpy as np

def num_days(x, e1_col1, e2_col1, e1_col2, e2_col2):
    
    e1_st = x[x['Event Name'] == 'Event 1'][e1_col1].reset_index(drop = True)
    e2_st = x[x['Event Name'] == 'Event 2'][e2_col1].reset_index(drop = True)
    e1_end = x[x['Event Name'] == 'Event 1'][e1_col2].reset_index(drop = True)
    e2_end = x[x['Event Name'] == 'Event 2'][e2_col2].reset_index(drop = True)
    
    
    if any(e1_st < e2_st) and any(e1_end > e2_end):
        return (e2_end - e2_st).tolist()[0].days
    
    elif any(e1_st > e2_st) and any(e1_end < e2_end):
        return (e1_end - e1_st).tolist()[0].days
    
    elif any(e1_st == e2_st) and any(e1_end == e2_end):
        return (e1_end - e1_st).tolist()[0].days
    
    elif any(e1_st <= e2_st) and any(e1_end <= e2_end):
        if any(e2_st >= e1_st) and any(e2_st <= e1_end):
            return (e1_end - e2_st).tolist()[0].days
    
        elif any(e1_end == e2_st):
            return 1
        
        else:
            return 0
    
    elif any(e2_st <= e1_st) and any(e2_end <= e1_end):
        
        if any(e1_st >= e2_st) and any(e1_st <= e2_end):
            return (e2_end - e1_st).tolist()[0].days
        
        elif any(e1_st == e2_end):
            return 1
        
        else:
            return 0
    else:
        pass   
```

```python
# Problem 3

import pandas as pd
import numpy as np

med = pd.read_excel("all_test_data_2023_10_16.xlsx", sheet_name = "medical_events")
enroll = pd.read_excel("all_test_data_2023_10_16.xlsx", sheet_name = "enrollment")
death = pd.read_excel("all_test_data_2023_10_16.xlsx", sheet_name = "death_dates")
```

```python
# Data frame merged

df = med.merge(enroll, how = 'left', on = 'patient_id').merge(death, how = 'left', on = 'patient_id')

df['event_start_date'] = pd.to_datetime(df['event_start_date'])
df['event_end_date'] = pd.to_datetime(df['event_end_date'])
df['enrollment_start_year_month'] = pd.to_datetime(df['enrollment_start_year_month'])
df['enrollment_end_year_month'] = pd.to_datetime(df['enrollment_end_year_month'])
df['death_date'] = pd.to_datetime(df['death_date'])
df['rank'] = df.groupby(['event_id'])['enrollment_start_year_month'].rank(method = 'first')
```

```python
df.sort_values(['patient_id','event_id','event_start_date'])
```

```python
# Separate each enrollment for each event
event_1 = df[df['rank'] == 1]
```

```python
# Separate each enrollment for each event
event_2 = df[df['rank'] == 2][['patient_id','event_id','enrollment_start_year_month','enrollment_end_year_month']]
event_2.columns = ['patient_id','event_id','sec_enrollment_start_year_month','sec_enrollment_end_year_month']
```

```python
event_mrg = event_1.merge(event_2, how = 'left', on =['patient_id','event_id'])
```

```python
# Modify the second question code

def enroll_days(x, eID, e1_col1, e2_col1, e1_col2, e2_col2):
    
    e1_st = x[x['event_id'] == eID][e1_col1].reset_index(drop = True)
    e2_st = x[x['event_id'] == eID][e2_col1].reset_index(drop = True)
    e1_end = x[x['event_id'] == eID][e1_col2].reset_index(drop = True)
    e2_end = x[x['event_id'] == eID][e2_col2].reset_index(drop = True)
    
    
    if any(e1_st < e2_st) and any(e1_end > e2_end):
        return (e2_end - e2_st).tolist()[0].days
    
    elif any(e1_st > e2_st) and any(e1_end < e2_end):
        return (e1_end - e1_st).tolist()[0].days
    
    elif any(e1_st == e2_st) and any(e1_end == e2_end):
        return (e1_end - e1_st).tolist()[0].days
    
    elif any(e1_st <= e2_st) and any(e1_end <= e2_end):
        if any(e2_st >= e1_st) and any(e2_st <= e1_end):
            return (e1_end - e2_st).tolist()[0].days
    
        elif any(e1_end == e2_st):
            return 1
        
        else:
            return 0
    
    elif any(e2_st <= e1_st) and any(e2_end <= e1_end):
        
        if any(e1_st >= e2_st) and any(e1_st <= e2_end):
            return (e2_end - e1_st).tolist()[0].days
        
        elif any(e1_st == e2_end):
            return 1
        
        else:
            return 0
    else:
        pass   
```

```python
# Append a list for each enrollment

fir_enroll = []

for i in event_mrg['event_id']:
    fir_enroll.append(enroll_days(event_mrg, i, 'event_start_date', 'enrollment_start_year_month' ,'event_end_date', 'enrollment_end_year_month'))

sec_enroll = []

for i in event_mrg['event_id']:
    
    if enroll_days(event_mrg, i, 'event_start_date', 'sec_enrollment_start_year_month' ,'event_end_date', 'sec_enrollment_end_year_month') == None:
        sec_enroll.append(0)
    else:
        sec_enroll.append(enroll_days(event_mrg, i, 'event_start_date', 'sec_enrollment_start_year_month' ,'event_end_date', 'sec_enrollment_end_year_month'))

```

```python
result = pd.concat([event_mrg, pd.Series(fir_enroll), pd.Series(sec_enroll)], axis = 1).loc[:, ['patient_id','event_id','event_start_date','event_end_date','enrollment_start_year_month','enrollment_end_year_month','sec_enrollment_start_year_month','sec_enrollment_end_year_month','death_date',0,1]] 

result.rename(columns = {0: 'fir_enroll_coverage', 1: 'sec_enroll_coverage'}, inplace = True)

result
```

```python
# Total coverage days

result['Sum_coverage'] = result['fir_enroll_coverage'] + result['sec_enroll_coverage']

# Find day difference between event end and death date
day_diff = result['event_end_date'] - result['event_start_date']
                                                  
day_diff = day_diff.dt.days

# Create event total days
result['total_event_days'] = day_diff

# Death days after event
ddays = result['death_date'] - result['event_end_date']
ddays = ddays.dt.days

result['death_days_after_event']  = ddays

# Total event days + death days after event

result['total_event_death'] = result['total_event_days'] + np.where(result['death_days_after_event'].isnull(), 0, result['death_days_after_event'])

result
```
